# 3.1 Relplot

```python
sns.relplot(data=df, x='Absences',y='Salary',kind='scatter')
```

```python
sns.scatterplot(data=df, x='Absences',y='Salary')
```

```python
df['DateofHire']=pd.to_datetime(df['DateofHire'])
tabla_serie= df[['DateofHire','Salary']].groupby(by=['DateofHire']).count().reset_index()

plt.figure(figsize=(12,6))
sns.relplot(data=tabla_serie, x='DateofHire',y='Salary',kind='line')
```

```python
sns.lineplot(data=tabla_serie, x='DateofHire',y='Salary')
```
