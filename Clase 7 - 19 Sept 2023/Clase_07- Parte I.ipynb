```python
df_lluvias = pd.read_csv('pune_1965_to_2002.csv')
df_lluvias.head()
```

```python
df_lluvias.index = df_lluvias['Year']
df_lluvias = df_lluvias.drop('Year', axis='columns')
df_lluvias.head()
```

```python
x = df_lluvias.index
y = df_lluvias.sum(axis='columns')
y
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(10, 4))
ax.plot(x, y, label='Precipitaciones acumuladas',color='r',marker='o')
ax.plot(x, y-800, label='Pr',color='b')
ax.set_xlabel('Año')                  
ax.set_ylabel('Precipitación acumulada (mm.)')
ax.set_title('Precipitaciones acumuladas según el año')
ax.legend(loc='upper left') 
```

# Modificación de hiperparametros graficos

```python
fig, ax = plt.subplots(figsize=(12, 3))  
ax.plot(df_lluvias.index, df_lluvias['Jan'], label='Precipitaciones de enero')
ax.plot(df_lluvias.index, df_lluvias['Feb'], label='Precipitaciones de febrero', color='C1')
maximo_enero = df_lluvias['Jan'].max()
maximo_febrero =  df_lluvias['Feb'].max()
ax.axhline(maximo_enero, color='red', linestyle='--', alpha=0.5,    linewidth=3, label='Máxima de enero')
ax.axhline(maximo_febrero, color='red', linestyle=':', alpha=0.5, linewidth=3, label='Máxima de febrero')
ax.set_xlabel('Año')  
ax.set_ylabel('Precipitación (mm.)') 
ax.set_title('Precipitaciones de enero y febrero')     
ax.set_xlim(df_lluvias.index[0], df_lluvias.index[-1])
ax.legend()     
```

# Grafico 1: histograma para ver distribucion del salario

```python
#from IPython.display import display
```

```python
# Orientado a objetos
mpl.style.use('classic')
fig, ax = plt.subplots(figsize=(10,6))
ax.hist(x= data.Salary, bins=None, density=False, orientation='vertical')
ax.set_xlabel('Salario')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma de Salario')
#fig.show()
display(fig)
```

```python
# Orientado a estados
mpl.style.use('classic')
plt.figure(figsize=(10,6))
plt.hist(x= data.Salary, bins=None, density=False, orientation='vertical')
plt.xlabel('Salario')
plt.ylabel('Frecuencia')
plt.title('Histograma de Salario')
display(fig)
```

Se logra ver que la distribucion de los datos es asimetrica a derecha con la mayoria de valores en la parte inferior de la distribucion (valores tipicos entre 50000-100000). Se logra ver tambien que hay presencia de valores por encima de 2000000

# Grafico 2: Piechart para ver distribucion de empleados por departamento 

```python
data_2=data[['EmpID','Department']].groupby(by='Department').count()
data_2
```

```python
data_2.index
```

```python
fig,ax = plt.subplots(figsize =(10, 6))
ax.pie(data_2.EmpID, labels = data_2.index,autopct='%1.1f%%')
ax.set_title('Piechart de distribucion de departamento')
display(fig)
```

```python
plt.figure(figsize=(10,6))
plt.pie(data_2.EmpID, labels = data_2.index,autopct='%1.1f%%')
plt.title('Piechart de distribucion de departamento')
plt.show()
```
