```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
```

# Metricas regresion

$$MAE = \frac{1}{n} \sum_{i=1}^n |(y_i -\hat{y_i})|$$

$$MSE = \frac{1}{n} \sum_{i=1}^n (y_i -\hat{y_i})^2$$

$$RMSE = \frac{1}{n} \sum_{i=1}^n \sqrt{(y_i -\hat{y_i})^2}$$

$$MAPE= \frac{1}{n} \sum_{i=1}^n  |\frac{y_i -\hat{y_i}}{y_i}|*100$$

$$R^2 = 1 - \frac{\sum_i (y_i -\hat{y_i})^2}{\sum_i (y_i -\bar{y})^2}$$

```python
import numpy as np
y=np.random.rand(1000) 
y
```

```python
y2 = y +np.random.rand(1000) 
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(y, linewidth=0.5)
plt.plot(y2,linewidth=0.5)
```

```python
mae= mean_absolute_error(y, y2)
mape =mean_absolute_percentage_error(y, y2)
mse =mean_squared_error(y, y2)
rmse= np.sqrt(mse)
r2 = r2_score(y, y2)
print(f"MAE : {round(mae,2)} -- MAPE: {round(mape,2)} -- MSE: {round(mse,2)} -- RMSE: {round(rmse,2)}, -- R2 : {round(r2,2)}")
```
